######Opacity data######
The opacity database is compiled from a variety of sources, combining lab-based data and numerical modeling. See [Freedman et al. 2008](http://arxiv.org/abs/0706.2374) and [Freedman et al. 2014](http://arxiv.org/abs/1409.0026) for details. 

######Abundance data######
The abundances of molecular, atomic, and ionic species are calculated using the thermochemical equilibrium models of Fegley & Lodders ([1994](http://adsabs.harvard.edu/abs/1994Icar..110..117F), [1996](http://adsabs.harvard.edu/abs/1996ApJ...472L..37F)); [Lodders (1999)](http://adsabs.harvard.edu/abs/1999ApJ...519..793L); [Lodders & Fegley (2002)](http://adsabs.harvard.edu/abs/2002Icar..155..393L); Lodders ([2002](http://adsabs.harvard.edu/abs/2002ApJ...577..974L), [2003](http://adsabs.harvard.edu/abs/2003ApJ...591.1220L)); [Lodders & Fegley (2006)](http://arxiv.org/abs/astro-ph/0601381). We adopt solar composition elemental abundances from [Lodders 2003](http://adsabs.harvard.edu/abs/2003ApJ...591.1220L)). These models assume "rainout" chemistry, in which clouds form in layers, removing material from the atmosphere (see [Lodders & Fegley (2006)](http://arxiv.org/abs/astro-ph/0601381) Section 1.4 for an overview).
